# Le serie

Vogliamo dare un senso preciso a somme di infiniti addendi, diciamo $a_i$ con $i\in{\mathbb N}$, che chiameremo <mark>serie</mark> e indicheremo con

$\qquad\displaystyle\sum_{i=0}^\infty a_i$.

Formalmente definiamo 

$\qquad\displaystyle\sum_{i=0}^\infty a_i\ \ =\ \ \lim_{n\to\infty}\sum_{i=0}^n a_i$.

A parole diremo che la serie è il limite della <mark>successione delle somme parziali</mark>. Ovvero della successione

$\qquad\displaystyle s_n\ \ =\ \ \sum_{i=0}^n a_i$.

Se il limite è finito, diremo che la serie <mark>converge</mark> se è infinito diremo che <mark>diverge</mark> altrimenti (cioè se il limiote non esiste) diremo che è è <mark>inteterminata</mark>.

# La serie geometrica

Raramente si riesce a calcolare il valore di una serie convergente. Noi considereremo due casi fortunati. Il primo caso è quello è la serie geometrica.

Supponiamo $a_i=p^i$ per un qualche $p\in{\mathbb R}$. Proviamo a calcolare la successione delle somme parziali

$\qquad\displaystyle s_n
\ \ =\ \ 
\sum_{i=0}^n p^i$.

è chiaro che (separando l'ultimo termine della somma)

$\qquad\displaystyle s_{n+1}
\ \ =\ \ 
\bigg(\sum_{i=0}^n p^i\bigg) + p^{n+1}$.

$\qquad\displaystyle\phantom{s_{n+1}}
\ \ =\ \ 
s_n+ p^{n+1}$.

Abbiamo anche (separando lil primo termine della somma) 

$\qquad\displaystyle s_{n+1}
\ \ =\ \ 
1 + \sum_{i=1}^{n+1} p^i$.

$\qquad\displaystyle\phantom{s_{n+1}}
\ \ =\ \ 
1 + p\sum_{i=0}^{n} p^i$.

$\qquad\displaystyle\phantom{s_{n+1}}
\ \ =\ \ 
1 + p\,s_n$.

Abbiamo quindi che

$\qquad\displaystyle\left\{\begin{array}{lcl}
s_{n+1}&=&s_n+ p^{n+1}\\
s_{n+1}&=&1 + p\,s_n\\
\end{array}\right.$

e di conseguenza

$\qquad s_n+ p^{n+1}\ =\ 1 + p\,s_n$

Risolvendo per $s_n$ otteniamo 

$\qquad$<mark style="padding:2ex;">$s_n=\dfrac{1 -p^{n+1}}{1-p}$</mark> 

Quindi a seconda del valore di $p$ ptteniamo 



$\qquad\displaystyle\sum_{i=0}^\infty p^i
\ \ =\ \ 
\left\{\begin{array}{ll}
\textrm{converge a }\dfrac{1}{1-p}&-1<p<1\\
\textrm{diverge se}& 1\le p\\[1ex]
\textrm{indeterminata se}& p\le -1\\
\end{array}\right.$
