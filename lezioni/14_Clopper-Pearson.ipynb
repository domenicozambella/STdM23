# Intervallo di confidenza di una proporzione

Consideriamo v.a. $X_p\sim B(1,p)$ con $p$ ignota. Per stimare $p$ usiamo $n$ variabili indipendenti $X_{p,1},\dots,X_{p,n}$ distribuite come $X_p$ (ovvero, ripetiamo l'esperimento $n$ volte).
La media empirica (o campionaria) è la seguente variabile aleatoria 

$$\bar X_p\quad=\quad\dfrac1n\sum^n_{i=1}X_{p,i}$$

Sappiamo che ${\rm E}(\bar X_p)=p$, questo perché $n\bar X_p\sim B(n,p)$ ha valore atteso $np$.

Supponiamo di osservare $\bar x$.

Fissiamo un $\alpha$, la significatività che vogliamo ottenere. 
Alternativamente fissiamo <mark>livello di confidenza</mark>, $1-\alpha$. 

Consideriamo l'insieme dei $p$ che soddisfano le seguenti condizioni

$\Pr(\bar X_p\le\bar x)\le\alpha/2$ 
and 
$\Pr(\bar X_p\ge\bar x)\le\alpha/2$

(abbiamo diviso equamente la probabilità d'errore a destra e sinistra, troveremo quindi un intervallo centrato in $\bar x$).

Il calcolo di questo intervallo va fatto numericamente.

Ogni libreria statistica ha una funzione per questo calcolo.

Questo metodo di stimare l'intervallo di confidenza si chiama di <mark>Clopper–Pearson</mark>

Vedremo un altro metodo più avanti.

## Confronto con il test di ipotesi

Supponiamo che applicando il metodo di Clopper–Pearson troviamo l'intervallo di confidenza $[p_1,p_2]$.

Immaginiamo il seguente test di ipotesi:

$H_0:$ la proporzione è $p_2$

$H_A:$ la proporzione è $<p_2$

Supponiamo di misurare $\bar x$ e calcoliamo il $p$-valore (i.e. la significatività can cui possiamo rigettare l'ipotesi nulla)

$\Pr(\bar X_p\le \bar x | p=p_2)= \alpha/2$

Il ragionamento simmetrico vale per $p_1$.
