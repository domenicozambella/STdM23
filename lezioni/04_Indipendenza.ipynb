# Indipendenza stocastica

Due eventi $A$ e $B$ si dicono <mark>(stocasticamente) indipenenti</mark> se 

$\quad\Pr(A\cap B)\quad=\quad\Pr(A)\cdot \Pr(B).$

Il seguente fatto è facile da verificare: se $A$ e $B$ sono eventi probabilità non nulla allora sono indipendenti se e solo se $\Pr(A|B)=\Pr(A)$ se e solo se $\Pr(B|A)=\Pr(B)$.

Due variabili aleatorie discrete $X$ ed $Y$ si dicono <mark>(stocasticamente) indipenenti</mark> se per ogni $x$ e $y$

$\quad\Pr(X, Y =x,y)\quad=\quad\Pr(X=x)\cdot \Pr(Y=y).$

Nel caso di variabili aleatorie continue la condizione diventa

$\quad\Pr(X\le x\ \textrm{ and }\ Y\le y)\quad=\quad\Pr(X\le x)\cdot \Pr(Y\le y)$
