# Variabili multivariate

Una v.a. <mark>multivariata</mark> è una v.a. $X$ che prende valori in uno spazio vettoriale (noi useremo solo $\mathbb{R}^k$). 

Conviene rappresentare $X$ come un vettore di v.a. univariate $X=[X_1,\dots,X_k]$ (in generale, non indipendenti).

Nel caso di variabili continue la <mark>funzione di ripartizione</mark> (cumulative distribution function) si definisce come segue

$\displaystyle\ F(x_1,\dots,x_k) =\ \Pr(X_1\le x_1,\dots,X_k\le x_k)$.

Spesso la funzione di ripartizione si può rappresentare come l'integrale di una funzione $f(x_1,\dots,x_k)$ chiamata <mark>densità di probabilità</mark> (joint probabilty distribution)

$\displaystyle\ F(x_1,\dots,x_k) =\ \int_{-\infty}^{x_1}\cdots\int_{-\infty}^{x_k} f(x_1,\dots,x_k)\  {\rm d}x_1\dots{\rm d}x_k$.





## Marginal distribution

Se integriamo la densità di probabilità rispetto a tutti i possibili valori di $x_1,\dots,x_k$ otteniamo ovviamente $1$. Se integriamo rispetto ai possibili valori di tutte tranne una delle variabili, per esempio $x_2$ otteniamo una densità di probabilità

$\displaystyle\ f_2(x_2) =\ \int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty}\cdots\int_{-\infty}^{+\infty} f(x_1,\dots,x_k)\  {\rm d}x_1{\rm d}x_3\dots{\rm d}x_k$.

che si chiama la <mark>distribuzione marginale</mark> di $X_2$.

In modo simile si definisce la marginale di $X_{i_1},\dots,X_{i_h}$ per ogni sottoinsieme di $X_1,\dots,X_k$.

È utile sapere che se $X_1,\dots,X_k$ sono indipendenti (cioè $X_1$ è indipendente da $[X_2,\dots,X_k]$ tutte le altre, e così per tutte le altre) allora $f(x_1,\dots,x_k)\ =\ f_1(x_1)\cdots f_n(x_n)$.

## Valore atteso

Il valore atteso di $X=[X_1,\dots,X_k]$ è per definizione ${\rm E}(X)=[{\rm E}X_1,\dots,{\rm E}X_k]$.

Nel caso di una v.a. continua con densità di probabilità $f(x_1,\dots,x_k)$ 

$\displaystyle{\rm E}(X_i)\ = \ \int_{-\infty}^{+\infty}\cdots\int_{-\infty}^{+\infty} x_i f(x_1,\dots,x_k)\  {\rm d}x_1\dots{\rm d}x_k$

oppure

$\displaystyle\phantom{{\rm E}(X_i)}\ = \ \int_{-\infty}^{+\infty} f_i(x_i)\  {\rm d}x_i$

dove $f_i(x_i)$ è la densità di probabilità marginale di $X_i$.

## Matrice di covarianza

Siano $X,Y$ v.a. univariate la <mark>covarianza</mark> è definita come

${\rm Cov}(X,Y)\ =\ {\rm E}(XY) - {\rm E}(X){\rm E}(Y)$

Si noti che la se $X$ e $Y$ sono indipendenti allora ${\rm Cov}(X,Y)=0$.

Inoltre ${\rm Cov}(X,X)={\rm Var}(X)$.

Invece che generalizzare ${\rm Var}(X)$ al caso multidimensionale, è più utile generalizzare ${\rm Cov}(X,X)$.

Supponiamo ora che $X,Y$ siano due v.a. bivariate (per semplicità).
Diciamo $X=[X_1,X_2]$ e $Y=[Y_1,Y_2]$.

$${\rm Cov}(X,Y)\ =\ 
\left[
\begin{array}{cc}
{\rm Cov}(X_1,Y_1)& {\rm Cov}(X_1,Y_2)\\
{\rm Cov}(X_2,Y_1)& {\rm Cov}(X_2,Y_2)
\end{array}
\right]$$

Quindi 


$${\rm Cov}(X,X)\ =\ 
\left[
\begin{array}{cc}
{\rm Cov}(X_1,X_1)& {\rm Cov}(X_1,X_2)\\
{\rm Cov}(X_2,X_1)& {\rm Cov}(X_2,X_2)
\end{array}
\right]$$


ovvero


$$\phantom{{\rm Cov}(X,X)}\ =\ 
\left[
\begin{array}{cc}
{\rm Var}(X_1)& {\rm Cov}(X_1,X_2)\\
{\rm Cov}(X_2,X_1)& {\rm Var}(X_2)
\end{array}
\right].$$
