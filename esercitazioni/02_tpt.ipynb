***Esercizio 2***

The executive of a company uses either a train or a bus to reach his office with equal probability. The train is late $10\%$ of the times. The bus is late $20\%$ of the times. What is the probability that the executive reaches late the office?

***Soluzione***

$B$ evento il dirigente prende autobus

$T$ evento il dirigente prende treno

$Br$ evento autobus in ritardo

$T\!r$ evento treno in ritardo

$R$ evento il dirigente è in ritaro

Il teorema delle probabilità totali dice

$$\Pr(R) = \Pr(Tr)\cdot\Pr(T|T\!r) +  \Pr(Br)\cdot\Pr(B|Br)$$

È implicito nel testo della domanda che la scelta treno/autobus sia indipendente dal ritardo. Quindi 

$$\Pr(T|T\!r) = \Pr(T)$$

$$\Pr(B|Br) = \Pr(B)$$

Il teorema delle probabilità diventa

$$\Pr(R) = \Pr(T\!r)\cdot\Pr(T) +  \Pr(Br)\cdot\Pr(B)$$

$$\Pr(R) = 0.1\cdot (1/2) + 0.2\cdot(1/2) = 0.15$$






***Esercizio 3***

We flip one of two coins. The first coin is fair, the other is biased $2/3$ for tail. We will receive 2€ for tail and 1€ for head. The coins is choosen randomly. The probability to pick the unfair coin is $1/3$. Compute the expected value of our winnings.

***Soluzione***

Sia $X$ la variabile aleatoria *vincita*

Calcoliamo la distribuzione di probabilità usando il teorema delle probabilità totali.

$$\Pr(X=1) = \Pr(F)·\Pr(H|F) + \Pr(B)·\Pr(H|B) $$

$$\Pr(X=2) = \Pr(F)·\Pr(T|F) + \Pr(B)·\Pr(T|B) $$

Dove 

$T$ evento *tail*

$H$ evento *head*

$F$ evento moneta *fair*

$B$ evento moneta *biased*

$$\Pr(X=1) = (2/3)·(1/2) + (1/3)·(1/3) = 4/9 $$

$$\Pr(X=2) = (2/3)·(1/2) + (1/3)·(2/3) = 5/9 $$

Il valore atteso è

$${\rm E}(X) = 1·\Pr(X=1) + 2·\Pr(X=1) = 14/9 $$
